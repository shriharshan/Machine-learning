The concept of boosting is not actually a machine learning algorithm, it is a methodolgy applied to an existing machine learning algorithm, most commanly applied to the decision tree.

## AdaBoost

- AdaBoost(Adaptive Boosting) works by using an ensemble of **weak learners** and then combining them through the use of a weighted sum.
- AdaBoost adapts by using previously created **weak learners** in order to adjust misclassified instances for the next created **weak learners**

### What is a **Weak Learner**

- A weak model is a model that is too simple to perform well on its own.
- The weakest decision tree possible would be a stump, one node, and two leaves
- Unlike a single decision tree which fits all the data at once (fitting the data hard). AdaBoost aggregates multiple weak learners, allowing the overall **ensemble** model to learn slowly from the features.

```python
X = df.drop('class', axis=1)
y = df['class']
```

```python
# X.isnull().sum()
```

```python
X = pd.get_dummies(X, drop_first=True, dtype=int)
```

```python
X
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(X,y, test_size=0.15, random_state=101)
```

```python
from sklearn.ensemble import AdaBoostClassifier
```

```python
model = AdaBoostClassifier(n_estimators=1)
```

```python
model.fit(X_train, y_train)
```

```python
from sklearn.metrics import classification_report, confusion_matrix, accuracy_score
```

```python
predictions = model.predict(X_test)
```

```python
predictions
```

```python
print(classification_report(y_test, predictions))
```

```python
model.feature_importances_
```

```python
model.feature_importances_.argmax() # return the index of 1
```

```python
error_rates = []

for n in range(1, 96):
    model = AdaBoostClassifier(n_estimators=n)
    model.fit(X_train, y_train)
    preds = model.predict(X_test)

    err = 1 - accuracy_score(y_test, preds)

    error_rates.append(err)
```

```python
plt.plot(range(1, 96), error_rates)
```

```python
feats = pd.DataFrame(index=X.columns, data=model.feature_importances_, columns=['Importance'])
feats
```
