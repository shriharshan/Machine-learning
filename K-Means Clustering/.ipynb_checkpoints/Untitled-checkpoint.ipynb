```python
X = pd.get_dummies(df)
X
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
scaler = StandardScaler()
```

```python
scales_X = scaler.fit_transform(X)
```

```python
from sklearn.cluster import KMeans
```

```python
model = KMeans(n_clusters = 2)
```

```python
# Fit transform
# fit_transform for scalling is exactly analogous to fit_predict for unsupervised learning
cluster_labels = model.fit_predict(scales_X)
```

```python
cluster_labels
```

```python
X['Cluster'] = cluster_labels
```

```python
X
```

```python
plt.figure(figsize=(12,6), dpi = 200)
X.corr()['Cluster'].iloc[:-1].sort_values().plot(kind='bar') # it returns how it's correlated to its particular cluster
```

```python
ssd = []
for k in range(2,10):
    model = KMeans(n_clusters = k)
    model.fit(scales_X)
    
    
    ssd.append(model.inertia_) #SSd point --> cluster cneter
```

```python
ssd
```

```python
plt.plot(range(2,10),ssd,'o--')
```

```python
pd.Series(ssd)
```
