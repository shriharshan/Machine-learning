## Conditional filtering based on single feature

```python
df[df['total_bill'] > 40]
```

```python
df[df['sex'] == 'Male'].head()
```

## Coniditional filtering based on multiple feature

**AND &** --- BOTH COND NEED TO BE TRUE

**OR |** --- EITHER COND IS TRUE

```python
df[(df['total_bill'] > 30) & (df['sex'] == 'Male')].head()
```

```python
df[(df['day'] == 'Sun') | (df['day'] == 'Sat')].head()
```

```python
df[df['day'].isin(['Sun','Sat'])].head()
```
