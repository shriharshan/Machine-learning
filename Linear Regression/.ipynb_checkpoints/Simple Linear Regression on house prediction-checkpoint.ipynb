```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("./Data/Advertising.csv")
```

```python
df.head()
```

```python
df['total_spend'] = df['TV'] + df['radio'] + df['newspaper']
```

```python
df.head()
```

```python
sns.scatterplot(data=df,x='total_spend',y='sales')
```

```python
# y = mx + b
# y = B1x + B0
```

```python
X = df['total_spend']
y = df['sales']
```

```python
# polyfit() uses ordinary least squares(OLS) 
np.polyfit(X,y,deg=1)
```

```python
potential_spend = np.linspace(0,500,100)
```

```python
# y = B1x + B0
predicted_sales = 0.04868788*potential_spend+4.24302822
```

```python
sns.scatterplot(x='total_spend',y = 'sales',data=df)
plt.plot(potential_spend,predicted_sales,color='red')
```

```python
spend = 200
predicted_sales = 0.04868788*spend+4.24302822
```

```python
predicted_sales
```

# Taking 3rd degree polynomail degree

```python
np.polyfit(X,y,3)  # x^3, x^2, x, y-intercept
```

```python
potential_spend = np.linspace(0,500,100)
```

```python
predicted_sales =   3.07615033e-07*potential_spend**3 + -1.89392449e-04*potential_spend**2 + 8.20886302e-02*potential_spend**1 + 2.70495053e+00
```

```python
sns.scatterplot(x='total_spend',y = 'sales',data=df)
plt.plot(potential_spend,predicted_sales,color='red')
```

```python
sp = 200
pred_sales = 3.07615033e-07*sp**3+-1.89392449e-04*sp**2+8.20886302e-02*sp+2.70495053e+00
```
