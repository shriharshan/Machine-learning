```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('./Data/AMES_Final_DF.csv')
df.head()
```

```python
df.info()
```

```python
df.describe()
```

```python
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
```

```python
X = df.drop('SalePrice', axis=1)
y = df['SalePrice']
```

```python
X_train, X_test, y_train, y_test = train_test_split(
        X, y, test_size=0.10, random_state=101)
```

```python
scaler = StandardScaler()
```

```python
scaled_X_train = scaler.fit_transform(X_train)
scaled_X_test = scaler.transform(X_test)
```

```python
from sklearn.linear_model import ElasticNet
```

```python
base_elastic_model = ElasticNet()
```

```python
param_grid = {'alpha': [ 1, 5, 10, 50 ,100],
             'l1_ratio': [0.0,0.5, 0.7, 0.9, 1.0]}   # a value 0.0 is L2(ridge) and 1 is L1(lasso)
```

```python
from sklearn.model_selection import GridSearchCV
```

```python
grid_model = GridSearchCV(base_elastic_model, 
                          param_grid=param_grid,
                         scoring = 'neg_mean_squared_error',
                         cv = 5, verbose = 0)
```

```python
grid_model.fit(scaled_X_train, y_train)
```

```python
grid_results = pd.DataFrame(grid_model.cv_results_)
grid_results
```

```python
sns.lineplot(
    x="param_alpha",
    y="mean_test_score",
    data=grid_results,
)

plt.xlabel("Alpha")
plt.ylabel("Mean Test Score")
plt.title("ElasticNet Grid Search Results")
plt.show()
```

```python
sns.violinplot(
    x="param_alpha",
    y="mean_test_score",
    data=grid_results,
    showmeans=True,
)

plt.xlabel("Alpha")
plt.ylabel("Mean Test Score")
plt.title("ElasticNet Grid Search Results (Violin Plot)")
plt.show()
```

```python
grid_model.best_params_ # as l1_ratio is 1.0 it is highly suggesting for L1 regularization
```

```python
y_pred = grid_model.predict(scaled_X_test)
```

```python
from sklearn.metrics import mean_squared_error, mean_absolute_error
```

```python
mean_absolute_error(y_test, y_pred)
```

```python
np.sqrt(mean_squared_error(y_test, y_pred))
```

# using lasso

```python
from sklearn.linear_model import Lasso
```

```python
lasso = Lasso(max_iter=10000)
```

```python
param_grid = {'alpha': [50, 100, 120, 140, 150]}
```

```python
lasso_grid = GridSearchCV(lasso, param_grid, cv=5)
```

```python
lasso_grid.fit(scaled_X_train, y_train)
```

```python
grid_results = pd.DataFrame(lasso_grid.cv_results_)
grid_results
```

```python
sns.lineplot(
    x="param_alpha",
    y="mean_test_score",
    data=grid_results,
)

plt.xlabel("Alpha")
plt.ylabel("Mean Test Score")
plt.title("Lasso Grid Search Results")
plt.show()
```

```python
sns.violinplot(
    x="param_alpha",
    y="mean_test_score",
    data=grid_results,
    showmeans=True,
)

plt.xlabel("Alpha")
plt.ylabel("Mean Test Score")
plt.title("Lasso Grid Search Results (Violin Plot)")
plt.show()
```

```python
print("Lasso", lasso_grid.best_params_)
```

```python
y_pred_lasso = lasso_grid.predict(scaled_X_test)
```

```python
print("Lasso: ",mean_absolute_error(y_test, y_pred_lasso))
```

```python
print("Lasso: ",np.sqrt(mean_squared_error(y_test, y_pred_lasso)))
```
