```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
np.random.seed(42)
num_samples = 100
feature1 = np.random.rand(num_samples) * 10
feature2 = np.random.rand(num_samples) * 5
feature3 = np.random.rand(num_samples) * 8
target = 3 * feature1 + 2 * feature2 + 4 * feature3 + np.random.rand(num_samples) * 2
```

```python
feature1
```

```python
target
```

```python
# Stack features into a single matrix
features = np.column_stack((feature1, feature2, feature3))
features, features.shape
```

```python
# Add bias tern to the features
features_with_bias = np.c_[np.ones(num_samples), features]
features_with_bias, features_with_bias.shape
```

```python
# Gradient Descent parameters
learning_rate = 0.01
num_iterations = 1000
```

```python
features_with_bias.shape[0], features_with_bias.shape[1]
```

```python
# Initialize coefficients
coefficients = np.zeros(features_with_bias.shape[1])
coefficients, coefficients.shape # shape od (4,) means it's 1-D array with 4 elements in it.
```

# The Gradient Descent Formula
\[
\theta_j := \theta_j - \alpha \frac{1}{n} \sum_{i=1}^{n} (h_\theta(x^{(i)}) - y^{(i)}) \times x_j^{(i)}
\]

### Explanation:

The gradient descent algorithm aims to minimize a cost function \(J(\theta)\) iteratively. In the formula:

\[
\theta_j := \theta_j - \alpha \frac{1}{n} \sum_{i=1}^{n} (h_\theta(x^{(i)}) - y^{(i)}) \times x_j^{(i)}
\]

Where:
- \(\theta_j\) is the \(j^{th}\) parameter (or coefficient) to be updated.
- \(h_\theta(x^{(i)})\) is the predicted value for the \(i^{th}\) sample, calculated using the linear regression model:
  \(\theta_0x_0^{(i)} + \theta_1x_1^{(i)} + \ldots + \theta_mx_m^{(i)}\), where \(x_0^{(i)}\) is 1 (bias term) and \(x_j^{(i)}\) represents the \(j^{th}\) feature of the \(i^{th}\) sample.
- \(y^{(i)}\) is the actual target value for the \(i^{th}\) sample.
- \(\alpha\) (alpha) is the learning rate, determining the step size during each iteration.
- \(n\) is the number of samples in the dataset.
- \(x_j^{(i)}\) represents the \(j^{th}\) feature of the \(i^{th}\) sample.

The term \(h_\theta(x^{(i)}) - y^{(i)}\) represents the error (or residual) for the \(i^{th}\) sample. The formula calculates the weighted sum of the errors for all samples, with weights determined by the corresponding features.

In each iteration, the parameters \(\theta_j\) are updated simultaneously for \(j = 0\) to \(m\), where \(m\) is the number of features. The goal is to iteratively adjust the parameters to minimize the cost function \(J(\theta)\). This method is effective for linear regression and various other machine learning algorithms.


```python
# Gradient Algorithm
for _ in range(num_iterations):
    # Calculate predictions
    predictions = np.dot(features_with_bias, coefficients)
    # Calculate errors 
    errors = predictions - target
    # update coefficients using gradient descent
    gradient = 2 * np.dot(features_with_bias.T, errors) / num_samples
    coefficients -= learning_rate * gradient

# print coefficients
print("Coefficients (Intercpet, feature1, feature2, feature3): ", coefficients)
print("Predictions: ", predictions)
print("Error: ", errors)
```

```python
# Regression Line Plot
final_predictions = np.dot(features_with_bias, coefficients) # for calculating the last prediction values for visualizing
plt.figure(figsize=(8, 6))
sns.set(style="whitegrid")
plt.scatter(x = target, y = final_predictions, color='red', label='Predictions')
sns.scatterplot(x=target, y=target, color='blue', label='Target')
sns.lineplot(x=target, y=final_predictions, color='green', label='Regression Line')
plt.xlabel("Actual Target")
plt.ylabel("Predicted Target")
plt.title("Actual vs. Predicted Values with Regression Line")
plt.legend()
plt.show()
```
